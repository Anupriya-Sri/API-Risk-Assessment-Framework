```python
# Import dependencies
import pandas as pd
import openpyxl

from sklearn.utils import resample      # scikit-learn bootstrap

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load data from excel
api_df = pd.read_excel("../../data/processed/preprocessed_train.xlsx")
api_df.head()
```

```python
api_df.shape
```

```python
# prepare bootstrapped data
for i in range(1, 10):
    # 100 bootstrapped samples of size 50 each
    boot_df = resample(api_df, replace=True, n_samples=1316)
    api_df = pd.concat([api_df, boot_df])
api_df
```

```python
api_df.shape
```

```python
# Saving bootstrapped data to excel
api_df.to_excel("../../data/processed/bootstrap.xlsx", index = False)
```
